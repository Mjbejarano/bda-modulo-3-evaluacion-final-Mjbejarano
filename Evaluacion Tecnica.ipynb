### Visualización de datos con matplotlib y seaborn.*

```python

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('df_todo.csv',index_col=0)
```

```python
df.head()
```

```python
# Quiero cambiar los valores de los meses de mi df:

meses_map = {
    1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril',
    5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto',
    9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
}

# Reemplazar los valores numéricos por los nombres de los meses
df['Month'] = df['Month'].map(meses_map)

df
```

### . ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

```python


# Definir el orden correcto de los meses
orden_meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 
               'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

# Convertir la columna 'mes' en un tipo categórico con orden específico
df['Month'] = pd.Categorical(df['Month'], categories=orden_meses, ordered=True)

# Ordenar el DataFrame según el orden de los meses
df = df.sort_values('Month')

plt.figure(figsize=(10,5))
sns.barplot(x='Month', y='Flights Booked', data=df, palette='viridis')
plt.xticks(rotation=45)  
plt.xlabel('Mes')
plt.ylabel('Cantidad de Reservas')
plt.title('Distribución de Reservas por Mes')
plt.show()
```

### . ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente?

```python
correlation_matrix = df[["Distance", "Points Accumulated"]].corr()

plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación entre Distancia de Vuelos y Puntos Acumulados")
plt.show()
```

### . ¿Cuál es la distribución de los clientes por provincia o estado?

```python
df.columns
```

```python
province_distribution = df['Province'].value_counts().index
province_distribution 
```

```python
sns.countplot(x='Province', data=df, color = 'c', order=province_distribution)
plt.xticks(rotation=45)
plt.xlabel("Provincia")
plt.ylabel("Nº clientes")
plt.show()
```

### . ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?

```python
Salario_Promedio= df.groupby('Education')['Salary'].mean().reset_index()
Salario_Promedio
```

```python
sns.barplot(x = "Education", 
            y = "Salary", 
            data = Salario_Promedio,
            errorbar = None, 
            palette = "magma")
plt.suptitle("Salario Promedio por Nivel Educativo")
plt.tight_layout()
```

### ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?

```python
Proporcion_clientes = df['Loyalty Card'].value_counts().index
Proporcion_clientes
```

```python
sns.countplot(x='Loyalty Card', data=df, color = 'c', order=Proporcion_clientes)
plt.xticks(rotation=0)
plt.xlabel("Tarjeta de Fidelidad")
plt.ylabel("Clientes")
plt.show()
```

### ¿Cómo se distribuyen los clientes según su estado civil y género?

```python
df.columns
```

```python
estado = df['Marital Status'].value_counts().index
sns.countplot(x='Marital Status', data=df, color = 'c', order=estado, hue = 'Gender')
plt.xticks(rotation=0)
plt.xlabel("Estado Civil")
plt.ylabel("Clientes")
plt.show()
```
